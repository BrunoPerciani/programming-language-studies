```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Entradas
1. Receber o peso da primeira moeda (A).
2. Receber o peso da segunda moeda (B).
3. Receber o peso da terceira moeda (C).

-- Processamento:
4. Comparar A, B e C (A = B = C?).
4.1 Se verdadeiro, mostrar que não há moedas falsas. Se falso, realizar o próximo passo.
5. Comparar A, B e C (A != B != C?).
5.1 Se verdadiro, não é possível definir a moeda falsa. Se falso, realizar o próximo passo.
6. Comparar a moeda A com B (A = B?).
6.1 Se verdadeiro, então a moeda C é falsa. Se falso, realizar o próximo passo.
7. Comparar a moeda A com C (A = C?).
7.1 Se verdadeiro, então a moeda C é falsa. Se falso, realizar o próximo passo.
8. Se nenhuma condição for atendida, então a moeda A é falsa.

-- Saídas:
9. Se os três valores forem iguais, apresentar que não há moedas falsas.
9.1 Se os três valores forem diferentes, apresentar que não há como definir a moeda falsa.
9.2 Apresentar qual é a moeda com o peso diferente dentre as comparadas.
```

# PROBLEMA 02
Seu amigo lhe perguntou: “Você consegue criar um programa capaz de identificar o maior, o menor e o intermediário entre três valores reais”. Você respondeu: “Só se for agora”.

**SOLUCIONANDO O PROBLEMA 02**


```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos identificar o maior, o menor e o intermediário entre três valores reais recebidos
- Existem restrições? Não existem restrições.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Receber o primeiro valor real.
- Receber o segundo valor real.
- Receber o terceiro valor real.
- Comparar os três valores reais.
- Identificar qual o maior valor, o intermediário e o menor valor.
- Apresentar o maior valor, o intermediário e o menor valor.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Entradas:
1. Receber o primeiro valor real (V1).
2. Receber o segundo valor real (V2).
3. Receber o terceiro valor real (V3).

-- Processamento:
4. Comparar os três valores (V1 = V2 = V3?).
4.1 Se verdadeiro, não é possível realizar comparação entre os três. Se falso, realizar o próximo passo.
5. Comparar o primeiro valor com o segundo e terceiro (V1 > V2 e V1 > V3?). Se for falso, realizar o próximo passo.
5.1 Se verdadeiro, o valor um é o maior. Comparar o valor dois e valor três (V2 > V3?).
5.2 Se verdadeiro, o valor dois é intermediário e o valor três é o menor. Se falso, o valor dois é o menor e o valor três é o intermediário.
6. Comparar o segundo valor com o primeiro e terceiro (V2 > V1 e V2 > V3?). Se for falso, realizar o próximo passo.
6.1 Se verdadeiro, o valor dois é o maior. Comparar o valor um e valor três (V1 > V3?).
6.2 Se verdadeiro, o valor um é intermediário e o valor três é o menor. Se falso, o valor um é o menor e o valor três é o intermediário.
7. Comparar o terceiro valor com o primeiro e segundo (V3 > V1 e V3 > V2?). Se for falso, realizar o próximo passo.
7.1 Se verdadeiro, o valor três é o maior. Comparar o valor um e valor dois (V1 > V2?).
7.2 Se verdadeiro, o valor um é intermediário e o valor dois é o menor. Se falso, o valor um é o menor e o valor dois é o intermediário.

-- Saídas:
8. Se os três valores forem iguais, apresentar que não é possível compará-los.
8.1 Caso os três valores não forem iguais, apresentar o maior valor, o valor intermediário e o menor valor.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Entradas:
### Vamos receber os três valores inteiros para podermos compará-los posteriormente.
valorum = float(input("Insira o primeiro valor real: "))
valordois = float(input("Insira o segundo valor real: "))
valortres = float(input("Insira o terceiro valor real: "))

## Processamento:
### Vamos criar uma função para identificar a posição de cada valor recebido.
def compararvaloresreais(valorum, valordois, valortres):
#### Se os três valores recebidos forem iguais não será possível compará-los. Se falso, fazer a próxima comparação.
    if valorum == valordois == valortres:
        return "Os três valores são iguais, não é possível compará-los."
#### Comparar o primeiro valor com o segundo e terceiro.
    elif valorum >= valordois and valorum >= valortres:
#### Se verdadeiro definir o valor um como maior e seguir. Se falso, realizar a próxima comparação.
        maior = valorum
#### Comparar o segundo valor com o terceiro.
        if valordois >= valortres:
#### Se verdadeiro, o valor dois é o intermediário e o valor três o menor. Se for falso, o contrário acontece.
            intermediario = valordois
            menor = valortres
        else:
            intermediario = valortres
            menor = valordois
#### Comparar o segundo valor com o primeiro e terceiro.
    elif valordois >= valorum and valordois >= valortres:
#### Se verdadeiro definir o valor um como maior e seguir. Se falso, realizar a próxima comparação.
        maior = valordois
#### Comparar o primeiro valor com o terceiro.
        if valorum >= valortres:
#### Se verdadeiro, o valor um é o intermediário e o valor três o menor. Se for falso, o contrário acontece.
            intermediario = valorum
            menor = valortres
        else:
            intermediario = valortres
            menor = valorum
#### Caso as condições anteriores não sejam verdadeiras, definir o valor três como maior e seguir.
    else:
        maior = valortres
#### Comparar o primeiro valor com o segundo.
        if valorum >= valordois:
#### Se verdadeiro, o valor um é o intermediário e o valor dois o menor. Se for falso, o contrário acontece.
            intermediario = valorum
            menor = valordois
        else:
            intermediario = valordois
            menor = valorum
#### Retornar de acordo com a resposta correta as variáveis definidas.
    return maior, intermediario, menor

## Saídas:
### Vamos apresentar o resultado de acordo com o retorno da função.
#### Se os três valores forem iguais, apresentar que não é possível compará-los. Se falso, realizar o próximo passo.
if compararvaloresreais(valorum, valordois, valortres) == "Os três valores são iguais, não é possível compará-los.":
  print("Os três valores são iguais, não é possível compará-los.")
else:
### Vamos identificar qual é o valor maior, intermediário e menor de acordo com o retorno correto da função.
  maior, intermediario, menor = compararvaloresreais(valorum, valordois, valortres)
### Vamos apresentar o maior, o intermediário e o menor valor.
  print("O maior valor é " + str(maior) + ", o valor intermediário é " + str(intermediario) + " e o menor valor é " + str(menor) + ".")
```

**OBSERVAÇÕES**

Na resolução desse exercício não consideramos o cenário em que o usuário insere dois valores iguais e um diferente, pois não haveria resultados absurdamente errados no final. Em tese, não seria possível identificar o valor intermediário. Nesse caso, o programa continua identificando o valor maior e menor e define o valor repetido como intermediário. # PROBLEMA 03

Advinha? Inventaram um novo metodo de tarifação da energia consumida em certo país tropical. Você foi escolhido na sua equipe para compreender esse método e implementar uma solução. Vai encarar? Veja que encrenca:

• Para os primeiros 50 watts (> 0 e ≤ 50), deve ser cobrado R$ 0,34 reais por watt.

• Para os proximos 100 watts (> 50 e ≤ 150), deve ser cobrado em R$ 0,48 reais por watt.

• Para os watts excedentes (> 150), e cobrado a exorbitância de R$ 0,70 reais por watt.

Assim, se um consumidor apresentar o consumo C = 231, o valor a ser pago será (50 ∗ 0, 34) + (100 ∗ 0, 48) + (81 ∗
0, 70).

**SOLUCIONANDO O PROBLEMA 03**


```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos descobrir o valor total a ser cobrado de uma tarifa de energia dado o consumo de energia em watts.
- Existem restrições? Sim, cada faixa de consumo em watts tem um valor específico a ser tarifado que deve ser considerado.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Receber o valor de energia consumida em watts do usuário.
- Comparar o valor da energia consumida em watts recebida com a faixa de tarifa.
- Calcular o valor consumido em reais de acordo com a taxa da tarifa da faixa considerando os excedentes.
- Somar os valores de cada faixa em reais.
- Apresentar o valor total em reais (R$) do consumo de energia recebido.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Entradas:
1. Receber o valor de energia consumida em watts do usuário (Consumo).

-- Processamento:
2. Comparar o valor recebido com a primeira faixa de consumo. (Consumo <= 50?).
2.1 Se verdadeiro, calcular o valor da tarifa (Total = Consumo * 0.34). Se falso, realizar a próxima comparação.
3. Comparar o valor recebido com a primeira faixa de consumo. (50 > Consumo <= 150?).
3.1 Se verdadeiro, calcular o valor da tarifa (Total = (50 * 0.34) + ((Consumo - 50) * 0.48)). Se falso, realizar a próxima comparação.
4. Se nenhuma das condições anteriores verdadeira, calcular o valor da tarifa (Total = (50 * 0.34) + (100 * 0.48) + ((Consumo - 150) * 0.70)) para (Consumo > 150).

-- Saídas:
5. Apresentar o valor total em reais (R$) do consumo de energia recebido de acordo com o retorno correto.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Entradas:
consumo = float(input("Qual o valor do consumo de energia em watts? (Insira somente um número): "))

## Processamento:
### Vamos comparar o consumo recebido com a primeira faixa de consumo da tarifa.
if consumo <= 50:
#### Se for verdadeiro, será calculado a tarifa total. Se for falso, vai realizar a próxima comparação.
  total = consumo * 0.34
#### Vamos comparar o consumo recebido com a segunda faixa de consumo da tarifa.
elif consumo > 50 and consumo <= 150:
#### Se for verdadeiro, será calculado a tarifa total. Se for falso, vai realizar a próxima comparação.
  total = (50 * 0.34) + ((consumo - 50) * 0.48)
else:
#### Se nenhuma das condições acima for verdadeira, vamos considerar que a última faixa de consumo da tarifa é verdadeira e calcular a tarifa total.
  total = (50 * 0.34) + (100 * 0.48) + ((consumo - 150) * 0.70)

## Saídas:
### Vamos apresentar o valor total da tarifa em reais de acordo com o retorno correto.
print("O valor total da tarifa de energia é de R$ %.2f"% total + ".")
```
