# PROBLEMA 01
Você tem um amigo que participa da *Greenpeace Brasil*. Essa organização o ativista recebe doações para diferentes projetos. Contudo, eles somente tem as informações brutas das doações. Será que você consegue implementar uma solução que ofereça o valor total anual e o valor médio anual anual das doações a partir do valor das doações mensais? Considere que você receberá doze valores mensais como entrada.

**SOLUCIONANDO O PROBLEMA 01**

```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos desenvolver um projeto que monte uma solução para a classificação do valor total anual e o valor médio anual das doações a partir do valor de doações mensais.
- Existem restrições? Deve ser recebido apenas doze valores mensais.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Pedir ao usuário para inserir os valores das doações mensais doze vezes.
- Calcular o valor total anual das doações, somando os valores das doações mensais.
- Calcular o valor médio anual das doações, dividindo o valor total anual pelo número de meses.
- Apresentar o valor total do ano valor médio anual ao usuário como parte dos resultados.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
1. Criar uma variável para armazenar os valores recebidos das doações mensais. (Doações mensais)
2. Criar um contador de meses. (Mês)

-- Processamento e estrutura de repetição:
3. Receber o valor em reais das doações mensais até que o contador de meses seja doze. (Doação)
4. Armazenar o valor recebido na variável criada.
5. Adicionar uma unidade no contador de meses.
6. Calcular a média dividindo o valor total por doze para calcular a média mensal. (Média = Valor total / 12)

-- Saídas:
7. Apresentar o valor total recebido em doações em um ano.
8. Apresentar a média mensal das doações no período de um ano.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
### Vamos criar variáveis para armazenar o valor total e o contador de meses.
valortotal = 0
mes = 1

## Processamento e estrutura de repetição:
### Vamos criar uma estrutura de repetição para receber o valor em real das doações mensais.
while mes <= 12:
### Enquanto o contador de meses for menor ou igual a doze, vamos receber o valor em reais das doações usando dados como valores reais.
    valordoacao = float(input(f"Digite o valor da doação para o mês {mes}: "))
### Vamos somar o valor da doação recebida anteriormente na variável de valor total.
    valortotal += valordoacao
### Vamos criar um incremento no contador de meses e somar uma unidade no valor de mês para que o código se finalize.
    mes += 1

### Quanto a estrutura de repetição for falsa e for finalizada vamos dividir o valor somado por doze para obter a média de doações mensais e armazenar em uma nova variável.
mediamensal = valortotal / 12

## Saídas:
### Vamos imprimir o valor total recebido de doações em um ano, bem como, a média de doações mensais em real.
print("Seja bem-vindo à página de voluntários do Greenpeace Brasil!")
### Ambos valores finais devem ser apresentados com duas casas decimais pois estamos tratando de valores em dinheiro.
print("O valor total das doações no último ano é de %.2f"% valortotal + " reais (R$).")
print("O valor médio das doações no último ano por mês é de %.2f"% mediamensal + " reais (R$).")
print("Obrigado por ser voluntário nesse ano, até mais!")
```

# PROBLEMA 02
Lembra-se dos números primos? Pois é, esses carinhas são muito relevantes no nosso dia-a-dia e você nem imagina o motivo, né? Você consegue implementar uma solução capaz de determinar que um certo número inteiro informado pelo usuário é ou não primo? Lembre-se que o número um não é primo. Interrompa a operação se o usuário informar um número negativo.

**SOLUCIONANDO O PROBLEMA 02**


```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos implementar uma solução capaz de determinar que um certo número inteiro informado pelo usuário é ou não é primo.
- Existem restrições? Sim, a solução não permite o uso de uma biblioteca específica.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Receber um número inteiro do usuário.
- Verificar se o número é negativo.
- Verificar se o número é equivalente a zero.
- Verificar se o número é equivalente a um.
- Se nenhuma das condições for verdadeira, verificar se o número é primo.
- Apresentar o resultado adequado e repetir o processo até que um número negativo seja inserido.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Processamento e estrutura de repetição:
1. Receber o número inteiro positivo do usuário. (Número)
2. Verificar se o número recebido é negativo.
3. Verificar se o número recebido é igual à um ou zero.
4. Se nenhuma condição for verdadeira, verificar se é primo ou não.
4.1 Verificar a quantidade de divisores que o número possui.

-- Saídas:
4.2 Se o número de divisores for nulo, apresentar para o usuário que o número é primo.
4.3 Caso contrário, apresentar para o usuário o número não é primo.
5. Repetir o processo até que seja inserido um número negativo.

OBS.: A operação só será finalizada quando o usuário inserir um número não positivo.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Processamento e saídas:
### Vamos criar uma estrutura de repetição para verificar se um número inteiro positivo é primo ou não.
while True:
### Vamos receber um número inteiro do usuário e posteriormente verificar se ele é primo ou não.
    numero = int(input("Insira um número inteiro positivo para verificar se ele é primo ou não (Ou insira um número negativo para encerrar a operação): "))

#### Vamos verificar se o número inserido é negativo ou não. Se for, a operação vai ser encerrada.
    if numero < 0:
        print("Você inseriu um número negativo então a operação foi encerrada!")
        break

#### Se for positivo, vamos verificar se é igual à um ou zero. Se for, vamos informar ao usuário que esses números não são primos.
    elif numero == 0 or numero == 1:
        print("Os números zero e um não são primos. Tente novamente!")

#### Caso nenhuma condição acima for verdadeira, vamos inicializar uma variável para contar os divisores do número (Um número primo só pode ter dois divisores).
    else:
### Vamos inicializar uma variável para contar o número de divisores do número inserido.
        divisores = 0
### Vamos inicializar outra variável como um divisor potencial do número inserido.
        divisorpotencial = 2
### Vamos criar uma nova estrutura de repetição para verificar se o número inserido é primo ou não.
        while divisorpotencial < numero:
### Vamos verificar se o número inserido é divisível pelo divisor potencial sem deixar um resto. Se sim, encontramos um divisor a mais que um para o número e ele não é primo.
            if numero % divisorpotencial == 0:
### Se encontrarmos um divisor, vamos incrementar uma unidade na variável de divisores.
                divisores += 1
                break
            divisorpotencial += 1
### Se o número de divisores for nulo, o número é primo. Então vamos apresentar a resposta adequada para cada caso.
        if divisores == 0:
            print(f"O número {numero} é primo.")
        else:
            print(f"O número {numero} não é primo.")
```

# PROBLEMA 03
Voce já participou na escola de brincadeiras educativas com espelhos? Pois é,  essas brincadeiras eram utilizadas para explorar ideias relacionadas a ótica. Eu participei de uma que consistia em colocar uma sequência de números e vê-los invertidos, isto é, a sequência 1234 aparecia como 4321. Voce consegue implementar uma solução para inverter uma sequência de números positivos de até cinco dígitos (99999)? Você só pode usar propriedades do sistema decimal e não vale manipular o número com uma string. O número será informado pelo usuário.

```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos desenvolver uma solução que inverta uma sequência de números positivos com cinco casas cada.
- Existem restrições? Sim, podem ser usadas apenas propriedades do sistema decimal, sem a manipulação de strings ou uso de listas.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Receber um número de até cinco dígitos e positivo do usuário.
- Verificar se o número atende às condições propotas.
- Identificar o último digito do número recebido.
- Somar o último dígito identificado à uma variável.
- Remover o último número do digito inserido.
- Repetir o processo até que o número inserido seja nulo.
- Apresentar o número invertido para o usuário.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Entradas:
1. Receber um número de cinco dígitos e positivo do usuário. (Número)

-- Processamento:
2. Verificar se o número recebido atende às condições.
4. Criar uma variável para receber os números invertidos.
5. Identificar o último digito do número recebido. (Dígito = Número % 10)
5.1 Somar o último dígito identificado à uma variável. (Número invertido = (Número invertido * 10) + Dígito)
5.2 Remover o último número do digito inserido. (Número //= 10)
5.3 Repetir o processo até que o número inserido seja nulo.

-- Saídas:
6. Apresentar o número invertido para o usuário.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Entradas:
### Vamos receber um número de cinco dígitos e positivo do usuário.
numero = int(input("Digite um número positivo de até cinco dígitos: "))

## Processamento:
### Vamos verificar se o número é positivo e tem cinco digitos.
if numero < 0 or numero > 99999:
### Se o número inserido for inválido, vamos apresentar uma mensagem para o usuário.
    print("Número inválido. Por favor, insira um número positivo de até cinco dígitos.")
else:
### Caso contrário, vamos inicializar uma variável para armazenar o número invertido
    numeroinvertido = 0
### Vamos criar uma estrutura de repetição para inverter o número recebido pelo usuário.
    while numero > 0:
### Vamos obter o último dígito do número.
        digito = numero % 10
### Vamos adicionar o dígito obtido ao número invertido somando o dígito à variável número invertido.
        numeroinvertido = (numeroinvertido * 10) + digito
### Vamos remover o último dígito obtido do número original, fazendo uma divisão por dez e descartando o resto da divisão.
        numero //= 10
### Vamos repetir o processo até que os dígitos do número original sejam nulos.

## Saídas:
### Vamos apresentar o número invertido ao usuário.
print("O número inserido na sua forma invertida é " + str(numeroinvertido) + "!")
```
