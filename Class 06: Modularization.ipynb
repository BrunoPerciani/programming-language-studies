# PROBLEMA 01
Voce está trabalhando em um congresso científico da SBC (Sociedade Brasileira de Computação) e seu voluntariou para resolver (por um algorítimo) o seguinte problema: obter o valor total das inscrições e o número de inscritos considerando as inscrições realizadas. Sabe-se que o valor de cada inscrição é de R$ 1000,00, mas existem alguns descontos para as seguintes situações:

*   Filiados da SBC a mais de 3 anos (inclusive) tem 40% de desconto;

*   Filiados com menos de 3 tem 30% de desconto;

*   Estudantes de graduação ou pós-graduação tem valores de inscrição fixados em R$ 100,00.

Em cada inscrição, você receber a indicação de cada inscrito (se é filiado, quanto tempo é filiado ou se é estudante) pela entrada do usuário.

**SOLUCIONANDO O PROBLEMA 01**

```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos obter o valor total das inscrições  e o número  de inscritos considerando as inscrições realizadas.
- Existem restrições? Sim. Filiados a mais de três anos tem 40% de desconto, filiados com menos de três anos tem 30% de desconto e estudantes tem valores de inscrição fixados em cem reais.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Receber o tipo de filiação de cada filiado.
- Contar cada passagem como um filiado.
- Calcular o total em reais da inscrição de cada usuário e adicionar ao total.
- Repetir até que o usuário pare de preencher informações dos filiados.
- Apresentar o total de filiados da associação e o total arrecadado das inscrições.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Processamento e estruturas de repetição:
1. Receber a situação cadastral do filiado. (Comum, estudante ou nenhuma)
2. Verificar se a operação foi finalizada.
2.1 A cada interação válida, somar uma unidade no contador de inscritos.
3. Se não for verdadeiro e a situação cadastral for filiado comum, calcular o valor conforme anos da filiação.
3.1 Receber o anos de filiação.
3.1.1 Se os anos recebidos forem maior ou igual a três, aplicar 40% de desconto. (Anos >= 3)
3.1.2 Se os anos recebidos forem maior ou igual a três, aplicar 40% de desconto. (Anos < 3)
4. Se não for verdadeiro e a situação cadastral for filiado estudante, calcular o valor conforme o valor fixo.
5. Caso contrário, pedir ao usuário repetir a interação.
6. Adicionar o valor calculado no contador de valor de inscrição.
7. Repetir operação até que o usuário finalize-a.

- Saídas:
8. Apresentar o valor final arrecadado das inscrições e o número total de inscritos.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
### Vamos adicionar uma mensagem de boas-vindas ao usuário.
print("Seja bem-vindo à Sociedade Brasileira de Computação!")
### Vamos inciar uma variável para realizar a contagem do valor em reais total das inscrições.
totaldasinscricoes = 0
### Vamos inciar uma variável para realizar a contagem do número total de inscritos.
numeroinscritos = 0

## Processamento e estruturas de repetição:
### Vamos criar uma estrutura de repetição para solicitar informações necessárias do usuário.
while True:
### Vamos solicitar a situação cadastral do usuário, isto é, se ele é filiado comum ou filiado estudante.
    tipoinscricao = input("Qual é a situação cadastral do filiado? (Insira 'F' para filiados comuns, 'E' para filiados-estudantes ou 'N' para encerrar a contagem: ")

### Vamos sempre verificar se a contagem pode ser encerrada.
    if tipoinscricao == 'N':
        break

### A cada entrada do usuário recebida, vamos adicionar uma unidade na contagem de número total de inscritos e prosseguir para a contagem do valor.
    numeroinscritos += 1

### Vamos calcular o valor em reais da inscrição recebida pelo usuário.
#### Se a inscrição do filiado for do tipo comum, vamos perguntar quantos anos este está filiado na associação.
    if tipoinscricao == 'F':
### Vamos receber quantos anos ele está filiado à associação.
        anosfiliado = int(input("Quantos anos esse filiado está na associação? "))
#### Se o ano recebido for igual maior que três, vamos aplicar um desconto de 40% no valor da inscrição.
        if anosfiliado >= 3:
### O usuário pagara portanto, somente 60% do valor do total da inscrição.
            valorinscricao = 1000 * 0.6
#### Caso contrário, ou seja, o ano recebido for menor que três, vamos aplicar um desconto de 30% no valor da inscrição.
        else:
### O usuário pagara portanto, somente 70% do valor do total da inscrição.
            valorinscricao = 1000 * 0.7
#### Se a inscrição for do tipo estudante, vamos considerar um valor fixo da taxa de inscrição.
    elif tipoinscricao == 'E':
        valorinscricao = 100
    else:
      print("Você inseriu uma situação cadastral invalida, tente novamente!")

### Vamos adicionar o valor em reais calculado na variável de valor das inscrições.
    totaldasinscricoes += valorinscricao
### Essa estrutura de repetição será realizada até que o usuário insira o valor 'N' e encerre a operação.

## Saídas:
### Vamos apresentar o valor em reais total das inscrições.
print(f"O valor total arrecadado das inscrições é de {totaldasinscricoes:.2f} reais (R$).")
### Vamos apresentar o número total de inscritos.
print(f"O número total de inscritos é {numeroinscritos}.")
```

# PROBLEMA 02
Vamos explorar elementos mais elementares de animações? Veja o padrão de quatro transições apresentado abaixo. Você consegue pensar em uma solução algorítimica para representar tal padrão? Ah! você deve apresentar asterísco por asterísco. Não vale implementação *hard code, i.e*., que implementa as transições de modo fixo no programa!

**SOLUCIONANDO O PROBLEMA 02**


```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos determinar um algorítmo capaz de representar a sequência de asteríscos formando uma animação.
- Existem restrições? Não existem restrições.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Receber um valor de parâmetro do usuário.
- Criar uma função para definir o padrão sequencial de transição.
- Definir as regras de transição da animação.
- Executar a função criada para imprimir o padrão.
- Apresentar o padrão impresso ao usuário.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Entradas:
1. Receber um valor de parâmetro do usuário (N).

-- Processamento:
2. Criar uma função para definir as regras de impressão do padrão.
3. Gerar um padrão de transição de maneira crescente a partir de um range (1, N + 1).
4. Gerar um padrão de transição de maneira decrescente a partir de um range (N - 1, 1, -1).
4.1 Adicionar um incremento de menos um em 'N' a cada iteração.
5. Gerar um padrão de transição de maneira crescente a partir de um range. (1, N + 1).
6. Gerar um padrão de transição de maneira decrescente a partir de um range (N - 1, 0, -1).
6.1 Adicionar um incremento de menos um em 'N' a cada iteração.

-- Saídas:
7. Apresentar o padrão impresso ao usuário.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Entradas:
### Vamos receber do usuário um valor 'n' para o parâmetro usado na estrutura de repetição.
n = int(input("Insira um valor inteiro: "))
### Vamos inserir uma linha de espaço para organizar o terminal.
print()

## Processamento e saídas:
### Vamos criar uma função para imprimir os asteríscos da maneira que desejamos.
def padraotransicoes(n):

### Vamos gerar um padrão de transição de maneira crescente.
### Vamos iterar de 1 até 'n' (Incluindo 'n').
    for i in range(1, n + 1):
### Vamos imprimir um valor correspondente de asteríscos em cada linha conforme a iteração 'i'.
        print("*" * i)

### Vamos gerar um padrão de transição de maneira decrescente.
### Vamos iterar de 'n' até 0, adicionando um incremento negativo a cada iteração até que o range termine.
    for i in range(n - 1, 1, -1):
        print("*" * i)
#### Note que mudamos o range final de 0 para 1, para gerar o efeito da animação que queremos.

### Vamos repetir o padrão de transição de maneira crescente.
### Vamos iterar de 1 até 'n' (Incluindo 'n').
    for i in range(1, n + 1):
        print("*" * i)

### Vamos repetir o padrão de transição de maneira decrescente.
### Vamos iterar de 'n' até 0, adicionando um incremento negativo a cada iteração até que o range termine.
    for i in range(n - 1, 0, -1):
        print("*" * i)
#### Note que mantivemos o range final em 0, para gerar o efeito da animação que queremos.

### Vamos chamar a função criada no início para executar o conjunto de instruções na ordem definida.
print("Confira o padrão de transição criado!" )
print()
padraotransicoes(n)
```
