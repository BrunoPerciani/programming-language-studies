# PROBLEMA 01
Voce sabe quais são os inteiros entre 1000 e 5000 (inclusive) que possuem o maior número de divisores? Divisor é um importante conceito matemático que permite responder perguntas do dia-a-dia como: “Se em uma linha de produção industrial, certo tipo de manutenção é feita na máquina A a cada 3 dias, na máquina B a cada 4 dias, e na máquina C, a cada 6 dias. Se no dia 2 de janeiro foi feita a manutenção das três máquinas, qual será o próximo dia em que a manutenção ocorrerá nas três máquinas?”. Implemente uma solução capaz de identificar e apresentar qual o último inteiro (entre 1000 e 5000) que possui o maior número de divisores. Note que você deve criar um módulo capaz de retornar o número de divisores de um dado número de entrada.

**SOLUCIONANDO O PROBLEMA 01**

```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos implementar uma solução capaz de identificar e apresentar qual o número inteiro entre 1000 e 5000 que possui o maior numero de divisores e quantos são os divisores.
- Existem restrições? O intervalo deverá ser de 1000 até 5000 com estes inclusos.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Criar uma função para contar divisores de números inteiros.
- Iterar sobre os valores que foram restringidos no enunciado.
- Calcular o número de divisores de cada iteração a partir da função.
- Apresentar o número com maior número de divisores inteiros.
- Apresentar quantos divisores esse número tem.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Modularização:
1. Iterar entre um e um número inteiro rebecido.
2. Inicializar uma variável para contar divisores.
3. Se a divisão inteira desse número pelo valor da iteração for nula, soma-se um no contador de divisores.
4. Retorna o número de divisores. (Divisores)

-- Processamento e estruturas de repetição:
5. Inicializar uma variável para armazenar o número com mais divisores. (Número com mais divisores)
6. Inicializar uma variável para armazenar o número de divisores. (Número de divisores)
7. Iterar entre 1000 e 5000 usando a função criada.
8. Toda vez que o número de divisores retornado for maior que o número de divisores do momento, atualiza esse valor.
9. Repete até chegar no número com mais divisores.

-- Saídas:
10. Apresentar o número com mais divisores.
11. Apresentar o número de divisores desse número.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Modularização:
### Vamos criar uma função para contar os divisores de números.
def contardivisores(numero):
### Vamos inicializar um contador para representar o número dos divisores de um número.
    contador = 0
### Vamos iterar entre um e um número recebido na função.
    for i in range(1, numero + 1):
### Se a divisão inteira entre o número recebido na função e o valor iterado for zero, adicionamos uma unidade ao contador.
        if numero % i == 0:
            contador += 1
### Após iterar por todos os valores, retornamos o contador.
    return contador

## Processamento:
### Vamos inicializar uma variável para armazenar o número com maior número de divisores.
numerocommaisdivisores = None
### Vamos inicializar uma variável para armazenar o número de divisores desse número.
maiornumerodedivisores = 0

### Vamos iterar sobre os inteiros entre 1000 e 5000:
for numero in range(1000, 5001):
### Vamos chamar o módulo criado, onde o número de divisores será dado pela função.
    numdivisores = contardivisores(numero)
#### Se o número de divisores retornado pelo contador da função for maior que o maior número de divisores até então.
    if numdivisores > maiornumerodedivisores:
### Definimos o número de divisores atual como o maior número de divisores e o número iterado no momento como o número com mais divisores.
        maiornumerodedivisores = numdivisores
        numerocommaisdivisores = numero

## Saídas:
### Vamos apresentar o número inteiro entre os valores iterados com mais divisores e quantos divisores ele tem.
print(f"O último inteiro entre 1000 e 5000 com o maior número de divisores é {numerocommaisdivisores}, com um total de {maiornumerodedivisores} divisores!")
```

# PROBLEMA 02
Um determinado clube de futebol pretende classificar seus atletas em categorias que sao definidas dependendo da sua faixa etária, conforme a tabela abaixo:

1. De 05 a 10 - Infantil;
2. De 11 a 15 - Juvenil;
3. De 16 a 18 - Junior I;
4. De 19 a 20 - Junior II;
5. De 21 a 35 - Profissional;

Implemente um algorítmo que solicite a idade, o nome e o pesem quilogramas  de cada atleta e apresente, em seguida, o seu nome, a sua categoria. Note que um módulo que devera retorna a categoria a partir da idade recebida. Assuma que seu programa deve parar quando o usuário digitar valores menores ou iguais a 4 ou maiores iguais a 36 para a idade do atleta. Após finalizar a leitura dos dados de todos os atletas, o programa deve calcular e imprimir:

* A quantidade de atletas que estao em cada categoria;
* A porcentagem de todos os atletas (independentemente da categoria) com peso entre 15,0 e 30,5 kg.

**SOLUCIONANDO O PROBLEMA 02**


```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos implementar um programa que solicite o nome, idade e peso de cada atleta e apresente, em seguida, o seu nome e sua categoria. Ao final, deve apresentar a quantidade de atletas por categoria.
- Existem restrições? Sim, deve-se seguir a tabela com as idades respectivas de cada categoria.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Criar um módulo para definir a categoria do atleta cadastrado conforme a idade.
- Receber a idade do atleta cadastrado e classificar a categoria.
- Receber o nome e a idade em quilogramas.
- Apresentar o nome de a categoria do atleta cadastrado.
- Verificar se o peso do atleta está entre 15.0 e 30.5 quilogramas.
- Repetir o processo até que a contagem termine.
- Ao final, apresentar a quantidade de atletas por categoria e a porcentagem de atletas com peso entre 15.0 e 30.5 quilogramas.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Modularização:
1. Se a idade estiver entre 5 e 10, retornar categoria "Infantil". (5 <= Idade <= 10)
2. Se a idade estiver entre 11 e 15, retornar categoria "Juvenil". (11 <= Idade <= 15)
3. Se a idade estiver entre 16 e 18, retornar categoria "Junior I". (16 <= Idade <= 18)
4. Se a idade estiver entre 19 e 20, retornar categoria "Junior II". (19 <= Idade <= 20)
5. Se a idade estiver entre 21 e 35, retornar categoria "Profissional".(21 <= Idade <= 35)

-- Processamento e estruturas de repetição:
6. Inicializar variáveis para cada categoria e total de atletas.
7. Criar uma estrutura de repetição para receber a idade, o nome e o peso do atleta.
7.1 Classificar a categoria conforme a função e a idade recebida.
7.2 Apresentar o nome e a categoria do atleta cadastrado.
7.3 Verificar se o peso do atleta está entre 15.0 e 30.5 quilogramas.
7.4 Somar uma unidade na variável de total de atletas. (Total de atletas)
7.5 Repetir o processo até que o usuário cancele a operação.
8. Calcular a porcentagem dos atletas que tem peso entre 15.0 e 30.5 quilogramas.

-- Saídas:
9. Apresentar a quantidade de atletas por categoria.
10. Apresentar a porcentagem de atletas que tem peso entre 15.0 e 30.5 quilogramas.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Modularização:
### Vamos criar uma função para determinar a categoria do atleta conforme sua idade.
def determinarcategoria(idade):
### Se a idade estiver entre 5 e 10, retornar categoria "Infantil".
    if 5 <= idade <= 10:
        return "Infantil"
### Se a idade estiver entre 11 e 15, retornar categoria "Juvenil".
    elif 11 <= idade <= 15:
        return "Juvenil"
### Se a idade estiver entre 16 e 18, retornar categoria "Junior I".
    elif 16 <= idade <= 18:
        return "Júnior I"
### Se a idade estiver entre 19 e 20, retornar categoria "Junior II".
    elif 19 <= idade <= 20:
        return "Júnior II"
### Se a idade estiver entre 21 e 35, retornar categoria "Profissional".
    elif 21 <= idade <= 35:
        return "Profissional"
    else:
        return "Não classificado"

## Processamento e estruturas de repetição:
### Vamos inicializar as variáveis para contar o número de atletas em cada categoria.
infantil = juvenil = juniorum = juniordois = profissional = 0
### Vamos inicializar as variáveis para contar o número total de atletas e o número de atletas com peso entre 15.0 e 30.5 quilogramas.
totalatletas = totalpeso = 0
### Vamos criar uma mensagem de boas vindas.
print("Seja bem-vindo! Vamos fazer a contagem dos atletas?")
print()

### Vamos criar uma estrutura de repetição para receber a informação de cada atleta.
while True:
    idade = int(input("Digite a idade do atleta que será cadastrado agora: "))

### Verifica se o looping deve ser encerrado. Caso contrário, continua.
    if idade <= 4 or idade >= 36:
        break

### Vamos receber o nome do atleta que esta sendo cadastrado.
    nome = str(input("Insira o nome do atleta que está sendo cadastrado: "))
    peso = float(input(f"Digite o peso do atleta {nome} em quilogramas (Kg): "))

### Vamos chamar o módulo de idades criado para determinar a categoria do atleta cadastrado.
    categoria = determinarcategoria(idade)
### Vamos incrementar uma unidade no contador da categoria correspondente à idade do atleta.
    if categoria == "Infantil":
        infantil += 1
    elif categoria == "Juvenil":
        juvenil += 1
    elif categoria == "Júnior I":
        juniorum += 1
    elif categoria == "Júnior II":
        juniordois += 1
    elif categoria == "Profissional":
        profissional += 1

### Vamos apresentar o nome e a categoria do atleta cadastrado.
    print(f"O atleta {nome} foi cadastrado com sucesso na categoria {categoria}!")
    print("Cadastre outro atleta ou digite '0' para encerrar a contagem.")
    print()
### Vamos incrementar uma unidade o contador do número total de atletas.
    totalatletas += 1

    # Verifica se o peso do atleta está entre 15.0 e 30.5 quilogramas e incrementa o contador correspondente.
    if 15.0 <= peso <= 30.5:
        totalpeso += 1

### Vamos calcular a porcentagem de atletas com peso entre 15.0 e 30.5 quilogramas.
porcentagem = (totalpeso / totalatletas) * 100

## Saídas:
# Imprime o número de atletas em cada categoria respectiva.
print()
print(f"Computamos os dados de {totalatletas} atletas. Confira abaixo:")
print(f"• {infantil} atleta(s) na categoria Infantil.")
print(f"• {juvenil} atleta(s) na categoria Juvenil.")
print(f"• {juniorum} Atleta(s) na categoria Júnior I.")
print(f"• {juniordois} atleta(s) na categoria Júnior II.")
print(f"• {profissional} atleta(s) na categoria Profissional.")
print(f"A porcentagem de de atletas com peso entre 15.0 e 30.5 quilogramas é de {porcentagem:.2f}%.")
```
