# PROBLEMA 01
Matrizes são muito utilizadas para resolver vários problemas técnico-científicos. Por isso, essa estrutura é central em diferentes pacotes estatísticos, tais como R, Matlab, Octave e SAS. Contudo, a matriz reduz sua
utilidade quando esparsa, isto é, a maior parte de suas células esté zerada. Crie uma função que receba uma matriz de inteiros de 5 × 5 e, então, retorne se a mesma é esparsa ou não (*True* ou *False*, respectivamente). É considerada esparsa a matriz cuja a quantidade de células zeradas é igual ou superior a (Número linhas ∗ Número colunas)/2.

**SOLUCIONANDO O PROBLEMA 01**

```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos implementar um módulo que retorne se uma matriz 5 X 5 é esparsa ou não.
- Existem restrições? Sim, existe uma regra para considerar se a matriz é esparsa ou não.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Criar uma função para definir se a matriz é esparsa ou não.
- Receber uma matriz 5 X 5 do usuário.
- Verificar se a matriz inserida é esparsa ou não.
- Apresentar o resultado final para o usuário.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Modularização:
1. Criar um módulo para verificar se uma matriz é esparsa ou não.
2. Contar a quantidade de células zeradas na matriz. (Células zeradas)
3. Contar o total de células na matriz. (Total de células)
4. Retornar se a expressão (Células zeradas >= (Total de células)/2) é verdadeira (Esparsa) ou não (Não esparsa).

-- Entradas:
5. Receber as componentes de cada linha da matriz.
6. Organizar os valores recebidos em uma matriz.

-- Processamento e saídas:
7. Usar o módulo criado para verificar se a matriz inserida é esparsa ou não.
8. Apresentar o resultado final para o usuário.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Modularização:
### Vamos criar um módulo para verificar se uma matriz é esparsa ou não.
def esparsa(matrix):
### Vamos contar a quantidade de células zeradas na matriz.
    celulaszeradas = sum(row.count(0) for row in matrix)
### Vamos contar o total de células na matriz.
    totalcelulas = len(matrix) * len(matrix[0])
### Verifica se a quantidade de células zeradas é igual ou superior a (Número linhas ∗ Número colunas)/2
    return celulaszeradas >= totalcelulas / 2

## Entradas:
### Vamos criar a função principal do programa.
def main():
### Vamos receber do usuário os elementos de uma matriz 5 x 5.
    print("Vamos verificar se uma matriz 5 X 5 é esparsa ou não. Por favor insira as componentes de cada linha da matriz!")
    print()
    matrix = []
### Vamos receber as componentes da matriz em cada linha dela.
    for i in range(5):
        print(f"• Insira as componentes da matriz na linha {i+1} separados por um espaço simples:")
### Vamos mudar a formatação dos valores recebidos de modo a termos uma matriz.
        row = list(map(int, input().split()))
        matrix.append(row)

## Processamento e saídas:
### Verifica se a matriz recebida é esparsa ou não e imprime o resultado.
#### Se o resultado for retornado pelo módulo for verdadeiro, a função é esparsa.
    if esparsa(matrix):
      print()
      for row in matrix:
        print(row)
      print("• A matriz inserida acima é esparsa!")
#### Se o resultado for retornado pelo módulo for falso, a função não é esparsa.
    else:
      print()
      for row in matrix:
        print(row)
      print("• A matriz inserida acima não é esparsa!")

### Vamos chamar a função principal e executar o programa.
if __name__ == "__main__":
    main()
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Modularização:
1. Criar uma função para percorrer cada linha da matriz definida.
1.1 Se a linha tiver um índice ímpar, percorrer os elementos da linha da esquerda para a direita.
1.2 Se a linha tiver um índice par, percorrer os elementos da linha da direita para a esquerda.

-- Processamentos e saídas:
2. Definir as componentes de uma matriz 5 X 5.
3. Chamar a função principal para retornar os valores percorridos.
4. Apresentar o conteúdo da matriz conforme a função criada.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Modularização:
### Vamos criar uma função para percorrer a matriz definida.
def navegandonamatriz(matrix):
### Vamos criar uma estrutura de repetição para percorrer cada linha da matriz.
    for i, row in enumerate(matrix):
### Vamos verificar se a linha é ímpar ou par a partir do índice.
#### Se a linha por ímpar, imprime a linha da esquerda para a direita.
        if i % 2 == 0:
            for col in row:
                print(col, end=" ")
#### Se a linha por par, imprime a linha da direita para a esquerda.
        else:
            # Se a linha é ímpar, imprime invertida
            for col in reversed(row):
                print(col, end=" ")

## Processamento e saídas:
### Vamos definir as componentes de uma matriz 5 X 5.
matrix = [
    [12, 19, 35, 49, 1],
    [65, 34, 21, 7, 67],
    [35, 12, 2, 67, 86],
    [34, 53, 2, 90, 68],
    [57, 86, 2, 98, 66]
]
### Vamos chamar a função principal para apresentar o conteúdo da matriz.
print(f"• O conteúdo da matriz 'M' é:")
navegandonamatriz(matrix)
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Modularização:
1. Criar uma função para substituir valores nulos dado uma matriz.
2. Iterar sobre cada linha da matriz, verificando se a linha contém algum valor nulo.
3. Se verdadeiro, vamos imputar valores para substituir o nulo. Caso contrário, segue adiante.
3.1 Filtrar os valores que não são nulos.
3.2 Calcular a média dos valores que não são nulos.
3.3 Substituir o valor nulo pela média calculada acima.

-- Processamentos e saídas:
4. Definir as componentes de uma matriz 4 X 4.
5. Chamar a função principal para imputar os valores na matriz original.
6. Apresentar a matriz original e a nova matriz para o usuário.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Modularização:
### Vamos criar uma função para imputar zeros dado uma matriz.
def imputezeros(matrix):
### Vamos iterar sobre cada linha da matriz.
    for row in matrix:
### Verifica se a linha contém algum valor nulo.
        if 0 in row:
#### Se verdadeiro, vamos imputar valores para substituir o nulo. Caso contrário, segue adiante.
### Vamos filtrar os valores que não são nulos.
            valoresnaonulos = [value for value in row if value != 0]
### Vamos calcular a média dos valores que não são nulos.
            valormedio = sum(valoresnaonulos) / len(valoresnaonulos)
### Vamos substituir o valor nulo pela média calculada acima.
            row[row.index(0)] = valormedio
### Repete o processo até iterar sobre todas as linhas da matriz.

## Processamento e saídas:
### Vamos definir as componentes de uma matriz 4 X 4 considerando que cada linha pode ter até um valor nulo.
matrix = [
    [10, 30, 0, 20],
    [20, 0, 20, 50],
    [21, 1, 23, 10],
    [0, 20, 20, 20]
]

### Vamos apresentar a matriz original.
print("• A matriz 'M' original é:")
for row in matrix:
    print(row)

### Vamos chamar a função principal para substituir os valores nulos da matriz.
imputezeros(matrix)

### Vamos apresentar a matriz com as imputações corretas usando o módulo criado.
print()
print("• A matriz 'M' com as imputações corretas é:")
for row in matrix:
    print(row)
```
