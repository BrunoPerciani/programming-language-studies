# PROBLEMA 01
Uma fábrica de camisetas produz os tamanhos PP, P, M, G e GG, cada qual com um preço distinto. A dona dessa fábrica gostaria de saber o valor total de vendas a partir do produto da quantidade de camisetas. PP, P, M, G e GG vendidas e os respectivos preços de venda. Você consegue ajudá-la com uma implementação?

**SOLUCIONANDO O PROBLEMA 01**

```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos descobrir o valor total das vendas a partir do produto da quantidade de camisetas vendidas.
- Existem restrições? Sim, cada tamanho de camiseta tem um preço distinto.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Receber a quantidade de camisetas vendidas de cada tamanho.
- Receber o preço de cada tamanho de camiseta.
- Multiplicar o preço da camiseta PP pela quantidade de vendas desse tamanho de camiseta.
- Multiplicar o preço da camiseta P pela quantidade de vendas desse tamanho de camiseta.
- Multiplicar o preço da camiseta M pela quantidade de vendas desse tamanho de camiseta.
- Multiplicar o preço da camiseta G pela quantidade de vendas desse tamanho de camiseta.
- Multiplicar o preço da camiseta GG pela quantidade de vendas desse tamanho de camiseta.
- Somar os produtos de cada tamanho de camiseta.
- Mostrar o valor total das vendas.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Entradas
1. Receber a quantidade de camisetas vendidas de cada tamanho. (Vendas: PP, Vendas: P; Vendas: M; Vendas: G; Vendas: GG)
2. Receber o preço de cada tamanho de camiseta. (Preço: PP, Preço: P, Preço: M, Preço: G, Preço: GG)

-- Processamento
3. Multiplicar o preço da camiseta PP pela quantidade de vendas desse tamanho de camiseta. (Total: PP)
4. Multiplicar o preço da camiseta P pela quantidade de vendas desse tamanho de camiseta. (Total: P)
5. Multiplicar o preço da camiseta M pela quantidade de vendas desse tamanho de camiseta. (Total: M)
6. Multiplicar o preço da camiseta G pela quantidade de vendas desse tamanho de camiseta. (Total: G)
7. Multiplicar o preço da camiseta GG pela quantidade de vendas desse tamanho de camiseta. (Total: GG)
8. Somar os produtos de cada tamanho de camiseta. (ST = Total: PP + Total: P + Total: M + Total: G + Tptal: GG)

-- Saídas
9. Mostrar o valor total das vendas.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Entradas:
### Vamos receber os dados referentes ao número de vendas de cada tamanho de camiseta. Os dados serão recebidos em números inteiros!
vendaspp = int(input("Quantas camisetas do tamanho PP foram vendidas? "))
vendasp = int(input("Quantas camisetas do tamanho P foram vendidas? "))
vendasm = int(input("Quantas camisetas do tamanho M foram vendidas? "))
vendasg = int(input("Quantas camisetas do tamanho G foram vendidas? "))
vendasgg = int(input("Quantas camisetas do tamanho GG foram vendidas? "))
### Vamos receber os dados referentes ao valor em reais que a fábrica cobra para cada tamanho de camiseta. Os dados serão recebidos em números reais!
precopp = float(input("Qual o valor em reais (R$) de uma unidade de camiseta no tamanho PP? (Use um ponto no lugar das vírgulas!) "))
precop = float(input("Qual o valor em reais (R$) de uma unidade de camiseta no tamanho P? (Use um ponto no lugar das vírgulas!) "))
precom = float(input("Qual o valor em reais (R$) de uma unidade de camiseta no tamanho M? (Use um ponto no lugar das vírgulas!) "))
precog = float(input("Qual o valor em reais (R$) de uma unidade de camiseta no tamanho G? (Use um ponto no lugar das vírgulas!) "))
precogg = float(input("Qual o valor em reais (R$) de uma unidade de camiseta no tamanho GG? (Use um ponto no lugar das vírgulas!) "))

## Processamento:
### Vamos multiplicar o número de vendas ao preço de cada respectivo tamanho de camiseta e armazenar em uma variável própria para cada tipo.
totalpp = vendaspp * precopp
totalp = vendasp * precop
totalm = vendasm * precom
totalg = vendasg * precog
totalgg = vendasgg * precogg
### Vamos somar o valor obtido pelo produto em cada tamanho de camiseta, assim totalizando o valor total das vendas
somatotal = totalpp + totalp + totalm + totalg + totalgg

## Saídas:
### Vamos apresentar o valor total das vendas
print("O valor total das vendas é de " + str(somatotal), "reais (R$)!")
```

# PROBLEMA 02
Em uma agência do INSS, todos os servidores calculam para cada pessoa atendida as seguintes. informações: a idade em  anos, meses, semanas e dias. E, você já deve ter percebido que isso é chato e sujeito a vários erros. Vamos lá! Você consegue ajudá-los! Implemente um algoritmo que realize esse cálculo a partir do ano de nascimento de uma pessoa e o ano corrente. Ah! voce pode assumir que cada mês possui 30 dias com 4,5 semanas.

**SOLUCIONANDO O PROBLEMA 02**


```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos descobrir a idade de uma pessoa em anos, meses, semanas e dias a partir da data de nascimento e a data da consulta.
- Existem restrições? Não há restrições.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Receber a data de nascimento do usuário.
- Receber a data em que a consulta está sendo feita.
- Calcular a diferença entre as datas recebidas.
- Descobrir o valor da idade do usuário em anos.
- Descobrir o valor da idade do usuário em semanas.
- Descobrir o valor da idade do usuário em meses.
- Descobrir o valor da idade do usuário em anos.
- Apresentar o valor da idade em anos, meses, semanas e dias.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Entradas
1. Receber a data de nascimento do usuário. (Data de nascimento)
2. Receber a data corrente em que a consulta está sendo feita. (Data vigente)

-- Processamento
3. Calcular a diferença de dias entre a data vigente pela data de nascimento do usuário. (Data corrente - Data de nascimento)
4. Descobrir o valor da idade do usuário em anos divindo a diferença de dias por 365. (Diferença de dias / 365) ### Um ano tem trezentos e sessenta e cinco dias.
5. Descobrir o valor da idade do usuário em meses dividindo a diferença de dias por 30. (Diferença de dias / 30) ### Um mês tem trinta dias.
6. Descobrir o valor da idade do usuário em semanas dividindo a diferença de dias por 30. (Idade em semanas // 4.5) ### Uma semana tem sete dias.
7. Definir que a diferença de dias entre a a data vigente pela data de nascimento do usuário é a idade em dias. (Diferença de dias = Idade em dias)

-- Saídas
8. Apresentar o valor da idade encontrada em anos em números inteiros.
9. Apresentar o valor da idade encontrada em meses em números inteiros.
10. Apresentar o valor da idade encontrada em semanas em números inteiros.
11. Apresentar o valor da idade encontrada em dias em números inteiros.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
from datetime import datetime
## Entradas:
### Vamos usar uma função da biblioteca para definir um formato, receber o dia, mês e ano de nascimento do usuário consultado e armazenar em uma variável.
datadenascimento = input("Qual a data de nascimento do usuário? Siga o formato [DD/MM/AAAA]! ")
datadenascimento = datetime.strptime(datadenascimento, "%d/%m/%Y")
### Vamos usar uma função da biblioteca para receber a data vigente e armazenar em uma variável.
dataatual = datetime.now()

## Processamento:
### Vamos calcular a diferença de dias entre a data da consulta e a data de nascimento do usuário.
diferencadedias = dataatual - datadenascimento
### Usando relações matemáticas vamos descobrir o valor da idade do usuário em anos e armazenar em uma nova variável [01 ano = 365 dias].
idadeanos = diferencadedias.days // 365  ### Note que usamos barras duplas para que o cálculo considere o resultado inteiro da divisão!
### Usando relações matemáticas vamos descobrir o valor da idade do usuário em meses e armazenar em uma nova variável [01 mês = 30 dias].
idademeses = diferencadedias.days // 12  ### Note que usamos barras duplas para que o cálculo considere o resultado inteiro da divisão!
### Usando relações matemáticas vamos descobrir o valor da idade do usuário em semanas e armazenar em uma nova variável [01 semana = 7 dias].
idadesemanas = diferencadedias.days // 7 ### Note que usamos barras duplas para que o cálculo considere o resultado inteiro da divisão!
### Usando relações matemáticas vamos apenas definir que a diferença de dias calculada na primeira parte é a idade do usuário em dias!
idadedias = diferencadedias.days

## Saídas:
### Vamos apresentar o valor da idade em anos, meses, semanas e dias respectivamente somente em números inteiros.
print("A idade deste usuário do INSS em anos é:", idadeanos)
print("A idade deste usuário do INSS em meses é:", idademeses)
print("A idade deste usuário do INSS em semanas é:", idadesemanas)
print("A idade deste usuário do INSS em dias é:", idadedias)
```

**OBSERVAÇÕES**

Resolvendo esse exercício pela primeira vez tentamos descobrir a idade em meses, semanas e dias a partir da idade em anos. Porém, realizando um teste de mesa para verificar o programa, percebemos que isso não traria tanta acurácia ao código uma vez que uma pessoa que nasceu em 2004, como nós, pode ter 20 ou 19 anos dependendo do mês em que se faz a consulta. Portanto, visando tornar o código mais eficiente, decidimos usar a bilbioteca *datetime* para receber a data de nascimento do usuário e a data vigente. O problema é que seu amigo compra as proporções erradas dos ingredientes, exigindo a compra às pressas de algum deles. Ajude seu amigo! Crie um algorítmo
capaz de calcular e informar quanto ele deve comprar de cada ingrediente a partir da quantidade de porções de *Oswaldo Aranha* que ele deseja vender.

**SOLUCIONANDO O PROBLEMA 03**


```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos descobrir a quantidade correta de cada ingrediente necessárias para fazer o total de porções do prato desejado.
- Existem restrições? Cada ingrediente tem uma quantidade específica pré-determinada necessária para compor uma porção do prato.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Receber a quantidade de porções desejadas do prato.
- Calcular a quantidade de contra-filé necessário em gramas e transformar em quilos.
- Calcular a quantidade de alho frito necessário em gramas e transformar em quilos.
- Calcular a quantidade de arroz branco necessário em gramas e transformar em quilos.
- Calcular a quantidade de batata portuguesa necessária em gramas e transformar em quilos.
- Calcular a quantidade de farofa de ovos necessária em gramas e transformar em quilos.
- Calcular a quantidade de ovos necessários em unidades.
- Apresentar a quantidade em quilos de cada ingredientes para compor o número de porções do prato desejadas.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Entradas
1. Receber a quantidade de porções desejadas do prato (Porções)

-- Processamento
3. Calcular a quantidade de contra-filé necessário em gramas. (Massa de contra-filé para uma porção * Porções)
4. Transformar esse valor em quilos e armazenar em uma nova variável. (Contra-filé em gramas / 1000)
5. Calcular a quantidade de alho frito necessário em gramas. (Massa de alho frito para uma porção * Porções)
6. Transformar esse valor em quilos e armazenar em uma nova variável. (Alho frito em gramas / 1000)
7. Calcular a quantidade de arroz branco necessário em gramas. (Massa de arroz branco para uma porção * Porções)
8. Transformar esse valor em quilos e armazenar em uma nova variável. (Arroz branco em gramas / 1000)
9. Calcular a quantidade de batata portuguesa necessária em gramas. (Massa de batata portuguesa para uma porção * Porções)
10. Transformar esse valor em quilos e armazenar em uma nova variável. (Batata portuguesa em gramas / 1000)
11. Calcular a quantidade de farofa de ovos necessária em gramas. (Massa de uma porção de farofa de ovos * Porções)
12. Transformar esse valor em quilos e armazenar em uma nova variável. (Farofa de ovos em gramas / 1000)
13. Calcular quantos ovos em unidades serão necessários. (Ovos = Porções)

-- Saídas
14. Apresentar o valor da massa necessária de cada ingrediente para montar o número de porções desejadas.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Entradas:
### Vamos receber a quantidade de porções que são desejadas para serem vendidas. Os dados serão recebidos em números inteiros!
porcoes = int(input("Quantas porções de Oswaldo Aranha você deseja vender? "))

## Processamento:
### Vamos calcular a quantidade de contra-filé necessário em gramas. Depois vamos transformar essa quantidade em quilos e armazenar em uma nova variável.
contrafilegramas = porcoes * 150
### Vamos considerar a seguinte relação [1 quilo = 1000 gramas].
contrafilequilos = contrafilegramas / 1000
### Vamos calcular a quantidade de alho frito necessário em gramas. Depois vamos transformar essa quantidade em quilos e armazenar em uma nova variável.
alhofritogramas = porcoes * 20
### Vamos considerar a seguinte relação [1 quilo = 1000 gramas].
alhofritoquilos = alhofritogramas / 1000
### Vamos calcular a quantidade de arroz branco necessário em gramas. Depois vamos transformar essa quantidade em quilos e armazenar em uma nova variável.
arrozbrancogramas = porcoes * 1100
### Vamos considerar a seguinte relação [1 quilo = 1000 gramas].
arrozbrancoquilos = arrozbrancogramas / 1000
### Vamos calcular a quantidade de batata portuguesa necessário em gramas. Depois vamos transformar essa quantidade em quilos e armazenar em uma nova variável.
batataportuguesagramas = porcoes * 60
### Vamos considerar a seguinte relação [1 quilo = 1000 gramas].
batataportuguesaquilos = batataportuguesagramas / 1000
### Vamos calcular a quantidade de farofa de ovos necessário em gramas. Depois vamos transformar essa quantidade em quilos e armazenar em uma nova variável.
farofadeovosgramas = porcoes * 90
### Vamos considerar a seguinte relação [1 quilo = 1000 gramas].
farofadeovosquilos = farofadeovosgramas / 1000
### Como os ovos são dados em unidades vamos definir que a quantidade de ovos é equivalente à quantidade de porções desejadas.
ovos = porcoes

## Saídas:
### Vamos apresentar os valores de quantidades para cada ingrediente obtidos anteriormente.
print("Para vender " + str(porcoes) + " porções de Oswaldo Aranha, serão necessários os ingredientes abaixo:")
print("• " + str(contrafilequilos) + " quilos de contra-filé.")
print("• " + str(alhofritoquilos) + " quilos de alho frito.")
print("• " + str(arrozbrancoquilos) + " quilos de arroz branco.")
print("• " + str(batataportuguesaquilos) + " quilos de batata portuguesa.")
print("• " + str(farofadeovosquilos) + " quilos de farofa de ovos.")
print("• " + str(ovos) + " unidades de ovos.")
print("Boas compras!")
```

# CHALLENGE

Sua amiga tem um pequeno mercado e, volta-e-meia, sofre com o problema de falta de troco. Como você está atento às necessidades do comércio, você decidiu empreender! "Vou implementar uma solução algorítimica capaz de determinar quantas cédulas são necessárias para trocar um determinado valor inteiro", você pensou. Por exemplo, se o usuário passar R$ 50 reais, o seu programa deverá informar que são necessárias:

• 2 notas de R$ 20 reais.

• 1 notas de R$10 reais.

• 0 nota de R$5 reais.

• 0 notas de R$2 reais.

• 0 nota de R$1 real.

```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos descobrir a quantidade de cédulas corretas para retornar de troco ou troca dado um valor inteiro em reais.
- Existem restrições? Serão considerados apenas valores inteiros e o troco deve ser composto pelo menor número de cédulas possível [20, 10, 5, 2, 1].
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Receber o valor a ser trocado em reais do usuário.
- Calcular quantas notas de vinte reais serão necessárias.
- Calcular quantas notas de dez reais serão necessárias.
- Calcular quantas notas de cinco reais serão necessárias.
- Calcular quantas notas de dois reais serão necessárias.
- Calcular quantas notas de um real serão necessárias.
- Apresentar a quantidade de cédulas de cada tipo a serem devolvidas.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Entradas
1. Receber o valor a ser trocado em reais do usuário em números inteiros. (Valor)

-- Processamento
2. Calcular quantas notas de vinte reais serão necessárias dividindo o valor recebido por vinte e armazenando em uma variável. (Valor // 20)
3. Armazenar o resto da divisão inteira. (Resto1 = Valor % 20)
4. Calcular quantas notas de dez reais serão necessárias dividindo o resto anterior por dez e armazenando em uma variável. (Resto1 // 20)
5. Armazenar o resto da divisão inteira. (Resto2 = Resto1 % 10)
6. Calcular quantas notas de cinco reais serão necessárias dividindo o resto anterior por cinco e armazenando em uma variável. (Resto2 // 5)
7. Armazenar o resto da divisão inteira. (Resto3 = Resto2 % 5)
8. Calcular quantas notas de dois reais serão necessárias dividindo o resto anterior por dois e armazenando em uma variável. (Resto3 // 2)
9. Armazenar o resto da divisão inteira. (Resto4 = Resto3 % 2)
10. Calcular quantas notas de um real serão necessárias dividindo o resto anterior por um e armazenando em uma variável. (Resto4 // 1)

-- Saídas
11. Apresentar a quantidade de cédulas de cada tipo a serem devolvidas em forma de lista.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Entradas:
### Vamos receber o valor inteiro em reais que o cliente quer trocar e armazenar em uma variável.
valor = int(input("Qual o valor inteiro em reais (R$) será trocado? "))

## Processamento:
### Vamos calcular quantas notas de vinte reais serão necessárias.
cedulasdevinte = valor // 20
### Vamos armazenar o resto da divisão usando o operador aritmético "%" e armazenar em uma nova variável.
resto1 = valor % 20
### Vamos calcular quantas notas de dez reais serão necessárias. As barras duplas representam uma divisão que só considera a parte inteira do quociente.
cedulasdedez = resto1 // 10
### Vamos armazenar o resto da divisão usando o operador aritmético "%" e armazenar em uma nova variável.
resto2 = resto1 % 10
### Vamos calcular quantas notas de cinco reais serão necessárias. As barras duplas representam uma divisão que só considera a parte inteira do quociente.
cedulasdecinco = resto2 // 5
### Vamos armazenar o resto da divisão usando o operador aritmético "%" e armazenar em uma nova variável.
resto3 = resto2 % 5
### Vamos calcular quantas notas de dois reais serão necessárias. As barras duplas representam uma divisão que só considera a parte inteira do quociente.
cedulasdedois = resto3 // 2
### Vamos armazenar o resto da divisão usando o operador aritmético "%" e armazenar em uma nova variável.
resto4 = resto3 % 2
### Vamos calcular quantas notas de um real serão necessárias. As barras duplas representam uma divisão que só considera a parte inteira do quociente.
cedulasdeum = resto4 // 1

## Saídas:
### Vamos apresentar a quantidade de cédulas de cada valor necessárias para compor o troco do valor recebido inicialmente conforme as condições.
print("Para trocar " + str(valor) + " reais (R$), serão necessárias: ")
print("• " + str(cedulasdevinte) + " cédulas de vinte reais (R$ 20).")
print("• " + str(cedulasdedez) + " cédulas de dez reais (R$ 10).")
print("• " + str(cedulasdecinco) + " cédulas de cinco reais (R$ 5).")
print("• " + str(cedulasdedois) + " cédulas de dois reais (R$ 2).")
print("• " + str(cedulasdeum) + " cédulas de um real (R$ 1).")
```
