```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Processamento e saídas:
1. Criar e imprimir o tabuleiro do jogador.
2. Criar e imprimir o tabuleiro do computador.
3. Posicionar as embarcações corretamente no tabuleiro do jogador.
4. Posicionar as embarcações corretamente no tabuleiro do computador e ocultar para o jogador humano.

5. Realizar o turno do jogador humano:
5.1 Solicitar que o jogador escolha uma linha e uma coluna para realizar um tiro no tabuleiro do computador. [Linha; Coluna]
5.2 Avaliar se o tiro acertou um navio ou se foi um tiro na água.
5.3 Se o tiro acertou um navio, marcar o acerto no tabuleiro do oponente. Se não, marcar a coordenada na água. ("X ou 0")
5.4 Verificar se todas as embarcações do oponente foram destruídas.
5.4.1 Se todas as embarcações foram destruídas, imprimir uma mensagem de vitória para o jogador e encerrar o jogo.
5.5 Se não todas as embarcações do oponente foram destruídas, imprimir o tabuleiro do computador atualizado.
5.6 Realizar o turno do computador:
5.7 Gerar aleatoriamente uma linha e uma coluna para o tiro do computador no tabuleiro do jogador. [Linha; Coluna]
5.8 Avaliar se o tiro do computador acertou um navio ou se foi um tiro na água.
5.9 Se o tiro do computador acertou um navio, marcar o acerto no tabuleiro do jogador. Se não, marcar a coordenada na água. ("X ou 0")
5.10 Verificar se todas as embarcações do jogador foram destruídas.
5.10.1 Se todas as embarcações do jogador foram destruídas, imprimir uma mensagem de derrota para o jogador humano e encerrar o jogo.
6. Se não todas as embarcações do jogador foram destruídas, repetir as rodadas até que o jogo se encerre.
OBS.: O jogo só irá parar quando todas as embarcações de um dos jogadores forem destruídas.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
### Vamos importar uma biblioteca que nos permite usar funções para gerar números aleatórios.
import random

## Criando funções importantes para o jogo funcionar:
### Vamos criar uma função para criar um tabuleiro vazio.
def criar_tabuleiro():
### Vamos criar uma matriz 10 X 10 para representar um tabuleiro vazio.
    tabuleiro = []
    for _ in range(10):
### Vamos definir que "~" representa a água.
        linha = ["~"] * 10
        tabuleiro.append(linha)
### Retorna o tabuleiro vazio e o armazena.
    return tabuleiro

### Vamos criar uma função para imprimir o tabuleiro criado anteriormente.
def imprimir_tabuleiro(tabuleiro):
    for linha in tabuleiro:
        print(" ".join(linha))

### Vamos criar uma função para criar e posicionar as embarcações no tabuleiro aleatoriamente.
def posicionar_embarcacoes(tabuleiro):
### Vamos criar um dicionário que irá definir as embarcações e quantas coordenadas cada embarcação ocupa.
    embarcacoes = {"Porta-aviões": 5, "Cruzador": 4, "Contratorpedeiro1": 3, "Contratorpedeiro2": 3, "Contratorpedeiro3": 3, "Rebocador1": 2, "Rebocador2": 2}
### Vamos criar uma função dentro da anterior para verificar se há espaço suficiente de acordo com as direções permitidas.
    def verifica_espaco(linha, coluna, tamanho, orientacao):
#### Se a orientação for horizontal, faz as conferências necessárias.
        if orientacao == "horizontal":
            for i in range(tamanho):
                if coluna + i >= 10 or tabuleiro[linha][coluna + i] != "~":
                    return False
                if linha > 0 and tabuleiro[linha - 1][coluna + i] != "~":
                    return False
                if linha < 9 and tabuleiro[linha + 1][coluna + i] != "~":
                    return False
#### Se a orientação for vertical, faz as conferências necessárias.
        else:
            for i in range(tamanho):
                if linha + i >= 10 or tabuleiro[linha + i][coluna] != "~":
                    return False
                if coluna > 0 and tabuleiro[linha + i][coluna - 1] != "~":
                    return False
                if coluna < 9 and tabuleiro[linha + i][coluna + 1] != "~":
                    return False
        return True

### Usando a biblioteca, vamos usar uma estrutura de repetição para escolher orientações aleatórias para cada embarcação no tabuleiro.
    for embarcacao, tamanho in embarcacoes.items():
        while True:
### Escolhe uma orientação aleatória para cada embarcação no tabuleiro.
            orientacao = random.choice(["horizontal", "vertical"])
#### Se a orientação for horizontal, escolhe uma coordenada aleatórias para linha.
            if orientacao == "horizontal":
### Vamos escolher uma coordenada aleatória para linha.
                linha = random.randint(0, 9)
### Vamos escolher uma coordenada aleatória para coluna.
                coluna = random.randint(0, 9 - tamanho)
### Vamos verificar o espaço definido de acordo com a função criada anteriormente.
                if verifica_espaco(linha, coluna, tamanho, orientacao):
                    for i in range(tamanho):
                        tabuleiro[linha][coluna + i] = embarcacao[0]
                    break
### Se a orientação for vertical, escolhe uma coordenada aleatórias para linha.
            else:
### Vamos escolher uma coordenada aleatória para linha.
                linha = random.randint(0, 9 - tamanho)
### Vamos escolher uma coordenada aleatória para coluna.
                coluna = random.randint(0, 9)
### Vamos verificar o espaço definido de acordo com a função criada anteriormente.
                if verifica_espaco(linha, coluna, tamanho, orientacao):
                    for i in range(tamanho):
                        tabuleiro[linha + i][coluna] = embarcacao[0]
                    break

### Vamos criar uma função que irá definir uma estrutura de repetição para o turno do jogador humano.
def turno_jogador(tabuleiro):
    while True:
        try:
### Vamos solicitar que o jogador humano escolha uma linha da matriz do tabuleiro.
            linha = int(input("Escolha a linha (0-9): "))
### Vamos solicitar que o jogador humano escolha uma coluna da matriz do tabuleiro.
            coluna = int(input("Escolha a coluna (0-9): "))
### Vamos verificar se as coordenadas escolhidas estão dentro dos limites do tabuleiro.
#### Se for verdadeiro, irá retornar o valor da linha e coluna recebidos.
            if 0 <= linha <= 9 and 0 <= coluna <= 9:
                return linha, coluna
#### Se for falso, irá apresentar a escala de valores que devem ser escolhidos e repetir.
            else:
                print("Por favor, escolha valores entre 0 e 9.")
        except ValueError:
            print("Entrada inválida. Por favor, insira um número.")

### Vamos criar uma função que irá definir uma estrutura de repetição para o turno do computador.
def turno_computador(tabuleiro):
### Vamos escolher uma linha aleatória da matriz do tabuleiro.
    linha = random.randint(0, 9)
### Vamos escolher uma coluna aleatória da matriz do tabuleiro.
    coluna = random.randint(0, 9)
### Vamos retornar o valor da linha e coluna recebidos pelo computador.
    return linha, coluna

## Processamento e saídas:
### Vamos criar a função principal do jogo de batalha naval!
def jogar_batalha_naval():
### Vamos começar saudando o jogador humano.
    print("Bem-vindo ao jogo Batalha Naval!")
### Vamos chamar a função criada anteriomente para criar o tabuleiro do jogador humano e do computador.
    tabuleiro_jogador = criar_tabuleiro()
    tabuleiro_computador = criar_tabuleiro()
### Vamos chamar a função criada anteriormente para posicionar as embarcações nos tabuleiros do jogador e do computador.
    posicionar_embarcacoes(tabuleiro_jogador)
    posicionar_embarcacoes(tabuleiro_computador)

### Vamos criar uma estrutura de repetição que conterá todo o passo-a-passo do jogo de batalha naval.
    while True:
### Irá imprimir o tabuleiro do jogador e receber uma linha e coluna do usuário.
        print("\nSeu tabuleiro:")
        imprimir_tabuleiro(tabuleiro_jogador)
        linha, coluna = turno_jogador(tabuleiro_computador)
#### Se a coordenada (linha; coluna) inseridas forem diferentes do caracter que representa a água, irá considerar que acertou um alvo.
        if tabuleiro_computador[linha][coluna] != "~":
            print("Parabéns! Você acertou um alvo!")
#### Se considerou que acertou um alvo, irá armazenar a coordenada inserida com um "X" e apresentar.
            tabuleiro_computador[linha][coluna] = "X"
#### Caso contrário, irá informar que o jogador humano deu um tiro na água e seguirá.
        else:
            print("Tiro na água!")
### Vamos marcar a água no tabuleiro do computador.
            tabuleiro_computador[linha][coluna] = "0"  # Marca a água no tabuleiro do computador



### Vamos verificar se todas as embarcações foram destruídas. Se sim, apresenta que o jogo se encerrou e o jogador humano venceu.
        if all(all(celula != "~" for celula in linha) for linha in tabuleiro_computador):
            print("\nParabéns! Você destruiu todas as embarcações inimigas!")
            break

#### Se não for verdadeiro, irá realizar o o próximo turno.
        print("\nTabuleiro do computador:")
### Vamos sempre apresentar a matriz do tabuleiro do computador com as coordenadas das embarcações ocultas, até que essas coordenadas não tomem um tiro.
        tabuleiro_computador_oculto = [["X" if celula == "X" else "~" for celula in linha] for linha in tabuleiro_computador]
        imprimir_tabuleiro(tabuleiro_computador_oculto)
### Vamos definir o turno do computador e receber as coordenadas (linha, coluna) da função criada anteriormente.
        print("\nTurno do computador:")
        linha, coluna = turno_computador(tabuleiro_jogador)
#### Se a coordenada (linha; coluna) inseridas forem diferentes do caracter que representa a água, irá considerar que acertou um alvo.
        if tabuleiro_jogador[linha][coluna] != "~":
            print("O computador acertou um alvo!")
#### Se considerou que acertou um alvo, irá armazenar a coordenada inserida com um "X" e apresentar.
            tabuleiro_jogador[linha][coluna] = "X"
#### Caso contrário, irá informar que o computador deu um tiro na água e seguirá.
        else:
            print("O computador atirou na água!")
### Vamos marcar a água no tabuleiro do computador.
            tabuleiro_jogador[linha][coluna] = "0"

### Vamos verificar se todas as embarcações foram destruídas. Se sim, apresenta que o jogo se encerrou e o computador venceu.
        if all(all(celula != "~" for celula in linha) for linha in tabuleiro_jogador):
            print("\nO computador destruiu todas as suas embarcações!")
            break
#### Se não for verdadeiro, irá realizar o o próximo turno e repetir as instruções até que o jogo se encerre.

### Define a função principal do jogo e a chama quando o script for executado.
if __name__ == "__main__":
    jogar_batalha_naval()
```

Você já usou ZIP ou TAR? Então você já fez uso da compactação que representa um procedimento amplamente empregado para reduzir o tráfego de imagens, vídeo e áudio. Um dos métodos consisteem suprimir valores sucessivos iguais, como ilustrado na figura. Crie uma solução que analise um vetor A e apresente
as informações da compressão. Assuma que o vetor A possui 15 valores inteiros fixos a sua escolha.

```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos compactar um vetor qualquer conforme a repetição do valor de suas componentes.
- Existem restrições? Sim, o vetor deve ter quinze valores inteiros fixos.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Criar uma função que decomponha o vetor conforme as requisições.
- Definir os quinze valores do vetor.
- Realizar a compactação do vetor escolhido.
- Apresentar o vetor original e o vetor compactado.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Modularização:
1. Criar uma lista para receber os valores compactados.
2. Criar um contador de frequência de cada valor do vetor.
3. Iterar sobre cada valor do vetor.
3.1 Se o valor atual for igual o anterior, somar um no contador de frequência.
3.2 Caso contrário, armazenar o valor e a sua frequência na lista e reiniciar o contador para o próximo valor.

-- Processamento:
4. Definir as quinze componentes do vetor original.
5. Chamar a função criada para compactar o vetor original definido.

-- Saídas:
6. Apresentar o vetor original e o vetor compactado.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Modularização:
### Vamos criar uma função para fazer a compactação do vetor escolhido.
def compactacao(vetor):
### Vamos inicializar uma lista para armazenar a compactação.
    comprimido = []
### Vamos inicializar uma variável para contar a frequência de cada elemento.
    contador = 1
### Vamos iterar sobre cada componente do vetor.
    for i in range(1, len(vetor)):
### Vamos verificar se o elemento atual da iteração é igual ao anterior.
        if vetor[i] == vetor[i - 1]:
#### Se verdadeiro, vamos incrementar uma unidade no contador de frequência.
            contador += 1
        else:
#### Caso contrário, adiciona o elemento do vetor e a frequência na lista de compactação.
            comprimido.append((vetor[i - 1], contador))
### Vamos reiniciar o contador de frequência.
            contador = 1
### Vamos adicionar o último elemento do vetor e a frequência na lista de compactação.
    comprimido.append((vetor[-1], contador))
### Vamos retornar a lista de compactação.
    return comprimido

## Processamento:
### Vamos definir o nosso vetor original com valores fixos.
vetor = [30, 30, 30, 8, 8, 30, 5, 5, 5, 5, 20, 20, 6, 6, 6]
### Vamos chamar a função criada para retornar o valor do compactado do vetor original.
resultado = compactacao(vetor)

## Saídas:
### Vamos apresentar ao usuário as componentes do vetor original.
print("• O vetor (A) original é " + str(vetor) + ".")
### Vamos apresentar ao usuário as componentes do vetor comprimido com uma formatação especial.
print("• O vetor (A) compactado é [", end="")
for i, (elemento, contador) in enumerate(resultado):
    if i < len(resultado) - 1:
        print(f"({elemento}.{contador}),", end="")
    else:
        print(f"({elemento}.{contador})", end="")
print("].", end="")
```

• Letra minúscula: deve ser convertida para o primeiro caractere a esquerda (Isto é, + 1);

```python
# PASSO 01: COMPREENDER O PROBLEMA
Queremos decodificar um texto qualquer recebido pelo usuário conforme a ASCII.
- Existem restrições? Sim, existem regras que devem ser seguidas para a decodificação.
```

```python
# PASSO 02: DECOMPOR O PROBLEMA
- Criar uma função para decodificar uma certa string conforme a ASCII.
- Verificar cada caractere da string e retornar o caractere convertido conforme as regras.
- Receber uma string do usuário para ser decodificada.
- Chamar a função principal para decodificar a string recebida.
- Apresentar a string decodificada ao usuário.
```

```python
# PASSO 03: PROJETO DE ALGORÍTIMO
-- Modularização:
1. Cria uma lista para receber os valores decodificados.
2. Iterar sobre cada caractere recebido na string.
2.1 Verificar se é maiúsculo. Se for, somar três caracteres a direita da ASCII.
2.2 Verificar se é minúsculo. Se for, somar um caractere a esquerda da ASCII.
2.3 Verificar se é "!". Se for, soma dois caracteres a direita da ASCII.
2.4 Verifica se é um caractere especial entre a posição 33 e 125. Se for, soma um caractere a esquerda da ASCII.
3. Repete o processo até iterar sobre todos os caracteres e retorna o resultado final na lista.

-- Entradas:
4. Receber uma string do usuário.

-- Processamento:
5. Chamar a função principal para decodificar a string.

-- Saídas:
6. Apresentar o resultado final da decodificação.
```

```python
# PASSO 04: IMPLEMENTAÇÃO DO PROGRAMA
## Modularização:
### Vamos criar uma função para decodificar uma certa string recebida do usuário.
def decodificador(string):
### Vamos iniciaizar uma lista para armazenar o resultado final.
    resultado = []
### Vamos iterar sobre cada caractere da string recebida.
    for char in string:
### Vamos verificar se o caractere da iteração atual é maiúsculo.
        if char.isupper():
#### Se verdadeiro, converte-o para o terceiro caractere da ASCII a direita. Caso contrário segue adiante.
            resultado.append(chr(ord(char) + 3))
### Vamos verificar se o caractere da iteração atual é minúsculo.
        elif char.islower():
#### Se verdadeiro, converte-o para o primeiro caractere da ASCII a esquerda. Caso contrário segue adiante.
            resultado.append(chr(ord(char) - 1))
### Vamos verificar se o caractere é "!".
        elif char == '!':
#### Se verdadeiro, converte-o para o segundo caractere da ASCII a direita. Caso contrário segue adiante.
            resultado.append(chr(ord(char) + 2))
### Vamos verificar se é um caractere especial entre a posição 33 e 125 da ASCII.
        elif 33 <= ord(char) <= 125:
#### Se verdadeiro, converte-o para o primeiro caractere da ASCII a esquerda. Caso contrário segue adiante.
            resultado.append(chr(ord(char) - 1))  # Converte para o primeiro caractere a esquerda
        else:
#### Caso nada for verdadeiro, mantém o caractere se não estiver dentro das regras estabelecidas.
            resultado.append(char)
### Vamos iterar sobre cada caractere da string recebida e retornar o resultado com a lista de caracteres codificados.
    return resultado

## Entradas:
### Vamos receber uma string do usuário para ser codificada.
texto = str(input("Digite um texto para ser codificado: "))

## Processamento:
### Vamos chamar a função criada para decodificar a string recebida do usuário.
resultado = decodificador(texto)

## Saídas:
### Vamos apresentar o resultado final da decodificação ao usuário como uma string.
print("• O texto inserido decodificado é: ")
print("".join(resultado))
```
